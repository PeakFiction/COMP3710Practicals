```python
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt

# Load dataset (will download first time)
lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4)

print("Dataset loaded:")
print("Images shape:", lfw_people.images.shape)   # (n_samples, height, width)
print("Data shape:", lfw_people.data.shape)       # (n_samples, n_features)
print("Number of classes:", len(lfw_people.target_names))

# Show first 12 images with labels
fig, axes = plt.subplots(3, 4, figsize=(12, 9))
for i, ax in enumerate(axes.flat):
    ax.imshow(lfw_people.images[i], cmap="gray")
    ax.set_title(lfw_people.target_names[lfw_people.target[i]])
    ax.axis("off")

plt.suptitle("Sample images from LFW dataset", fontsize=16)
plt.show()
```
